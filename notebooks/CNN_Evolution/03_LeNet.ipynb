## 1. History & Motivation
LeNet-5 is a seminal convolutional neural network designed by Yann LeCun in 1998.

## 2. Intuition & Architecture
It introduces the core concepts of weight sharing, local receptive fields, and sub-sampling.

## 3. Mathematical Details
Convolutions use shared weights: $O(x, y) = \sum_{i,j} I(x+i, y+j) W(i, j) + b$.

## 4. Layer-by-Layer Implementation
Below is the from-scratch implementation of LeNet-5.

```python
import torch
import torch.nn as nn

class LeNetScratch(nn.Module):
    def __init__(self, in_channels=3, num_classes=10):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv2d(in_channels, 6, 5),
            nn.Tanh(),
            nn.AvgPool2d(2, 2),
            nn.Conv2d(6, 16, 5),
            nn.Tanh(),
            nn.AvgPool2d(2, 2),
            nn.AdaptiveAvgPool2d((5, 5))
        )
        self.classifier = nn.Sequential(
            nn.Linear(16 * 5 * 5, 120),
            nn.Tanh(),
            nn.Linear(120, 84),
            nn.Tanh(),
            nn.Linear(84, num_classes)
        )
        
    def forward(self, x):
        x = self.features(x)
        x = torch.flatten(x, 1)
        return self.classifier(x)

print("LeNet successfully defined from scratch!")
```
