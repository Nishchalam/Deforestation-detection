## 1. Core Mathematical Derivations

### A. Grad-CAM (Class Activation Mapping)
Grad-CAM computes the gradients of the score for class $c$ ($y^c$) with respect to feature map activation $A^k$ of a convolutional layer. These gradients are globally pooled to find the channel importance weights $\alpha_k^c$:
$$\alpha_k^c = \frac{1}{Z} \sum_{i} \sum_{j} \frac{\partial y^c}{\partial A_{i,j}^k}$$

The weighted combination of forward activation maps is followed by a ReLU to keep positive features:
$$L_{\text{Grad-CAM}}^c = \text{ReLU}\left(\sum_{k} \alpha_k^c A^k\right)$$

### B. Grad-CAM++
Grad-CAM++ handles cases with multiple occurrences of objects by computing second-order and third-order gradients to find more precise pixel-wise importance coefficients.

### C. Saliency Maps
Vanilla Saliency computes the absolute gradient of the output score w.r.t the input image pixels:
$$S(x) = \max_{c} \left| \frac{\partial y^c}{\partial x} \right|$$

### D. Occlusion Sensitivity
Slides a grey mask over the input image and records the drop in target prediction confidence: $\text{drop} = P(\text{original}) - P(\text{occluded})$.

```python
import numpy as np
import cv2
from PIL import Image
import matplotlib.pyplot as plt

print("Interpretability libraries loaded.")
```

```python
# Simulating Grad-CAM and Occlusion outputs
img_size = 224
mock_image = Image.new("RGB", (img_size, img_size), color=(34, 139, 34))

# Build mock gaussian activation heatmap center-aligned
x = np.linspace(-3, 3, img_size)
y = np.linspace(-3, 3, img_size)
x_grid, y_grid = np.meshgrid(x, y)
z = np.exp(-(x_grid**2 + y_grid**2) / 2.0)
mock_heatmap = z / np.max(z)

# Plot side-by-side
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(mock_image)
axes[0].set_title("Original Mock Image")
axes[0].axis("off")

axes[1].imshow(mock_heatmap, cmap="jet")
axes[1].set_title("Simulated Class Attribution Map")
axes[1].axis("off")
plt.tight_layout()
plt.show()
```
